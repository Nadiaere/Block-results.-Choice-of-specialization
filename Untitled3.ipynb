###1.1. Скачать данные в закрепе к уроку (данные по недвижимости)

```python
df = pd.read_csv ('/content/kc_house_data.csv') 
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```


##1.2 Изучите стоимости недвижимости


Рассмотрим представленый график, с помощью линейного графика, можно увидеть как меняется стоимость жилья  в зависимости от количества комнат

```python
plt.figure(figsize=(10, 5))

sns.lineplot(x=df['bedrooms'], y=df['price'])

plt.title('Стоимость недвижимости')
plt.xlabel('Кол-во спален')
plt.ylabel('Стоимость');
```

## 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize = (10,5))

sns.histplot(df['sqft_living'], bins = 10)
plt.xlabel('Жилые квадратные метры')
plt.ylabel('Количество')
plt.title('Распределение жилой квадратуры');
```

Здесь мы видим распределении населения по размеру жилой площади или жилых квадратных метров

##1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(10, 5))

plt.plot(df['yr_built'])

plt.title('Распределение года постройки')
plt.xlabel('Количество')
plt.ylabel('Год постройки');
```

Здесь мы видим, что основное количество домов построены в период от 1900 до 2000 года, старых домой как и новых меньше

##Условие 2: 2 задача

###2.1 Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
view = df['waterfront'].value_counts()

plt.figure(figsize = (10,5))
plt.pie(view.values, autopct = '%1.1f%%')
plt.title('Распределение домов c видом на набережную')
plt.legend(['No', 'Yes']);
```

На этой круговой диаграмме предствлено наглядно, что свыше 99% , не имеют вида на набережную

## 2.2 Изучите распределение этажей домов

```python
floors = df['floors'].value_counts()

plt.figure(figsize = (10,5))
plt.pie(floors.values, autopct = '%1.1f%%', labels = floors.index)
plt.title('Распределение этажей домов');
```

Из диаграммы видно, что большая часть домов это малоэтажное строительство

##2.2 Изучите распределение состояния домов

```python
sns.boxplot(y='price', x='condition' , data=df)
```

Используя быструю оценку состояния домов, можно сделать вывод, что основное количество домов находятся в среднем состоянии и нет большого разрыва в стоимости

*   Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.



```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) <= 0.3] = 0
corr_matrix
```

посмотрим на тепловую карту корреляций:

```python
plt.figure(figsize = (15,10))
sns.heatmap(df.corr(),  cmap = 'viridis',annot = True)
```

```python
counts = df['grade'].value_counts()
    
plt.figure(figsize=(8, 6))    
plt.title('grade')
sns.barplot(x=counts.index, y=counts.values)
    
plt.show()
```

```python
sns.jointplot(x = df['price'], y = df['sqft_living'], kind = 'reg');
```

```python
plt.figure(figsize = (10,5))
sns.boxplot(x = df['price'], y = df['view'].astype('str'), whis = 1.5)
plt.xlabel('price')
plt.ylabel('view')
plt.title('Влияние вида на стоимость');
```
